```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the dataset
print("Please upload the 'Unemployment in India' dataset CSV file:")
uploaded = files.upload()

# Read the uploaded file
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Strip extra spaces from column names
df.columns = df.columns.str.strip()

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with invalid/missing 'Date' or 'Estimated Unemployment Rate (%)'
df = df.dropna(subset=['Date', 'Estimated Unemployment Rate (%)'])

# Ensure 'Estimated Unemployment Rate (%)' is numeric
df['Estimated Unemployment Rate (%)'] = pd.to_numeric(df['Estimated Unemployment Rate (%)'], errors='coerce')

# Display dataset info
print("\nDataset Info:")
print(df.info())

# Visualization 1: Unemployment rate over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'], marker='o', color='purple', label='Unemployment Rate')
plt.title('Unemployment Rate Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```
